Calcular el factorial de un número

```python
def factorial(num):
  if (num <= 1): return 1
  else: return factorial(num - 1) * num
```

```python
factorial(5)
```

Calcular la suma de una serie

```python
def suma_serie(min, max):
  if max <= min: return print("El valor de max tiene que ser mayor a min")
  resultado = 0
  for i in range(min, max + 1):
    resultado += i
  return resultado
```

```python
suma_serie(1, 3)
```
