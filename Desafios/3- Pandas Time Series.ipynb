```python
summer_olympic_archivo = pd.read_csv('C:/Users/Usuario/Documents/Coderhouse/Summer-Olympic-medals-1976-to-2008.csv', encoding='ISO-8859-1')
```

```python
summer_olympic_archivo_fill = summer_olympic_archivo.fillna(0)
```

```python
años_juegos_olimpicos = summer_olympic_archivo_fill['Year']
años_juegos_olimpicos
```

```python
summer_olympic_archivo['Year'] = pd.to_datetime(años_juegos_olimpicos, format = '%Y', errors ='ignore')
```

```python
summer_olympic_archivo['Year'] #ya queda nuestro año
```

```python
indice_año = summer_olympic_archivo.set_index('Year') #Lo seteamos como indice
```

```python
indice_año.head()
```

**2. Separar nombres de apellidos en dos columnas distintas usando la coma como separador**

```python
atletas_juegos_olimpicos = indice_año['Athlete']
atletas_juegos_olimpicos
```

```python
atletas_juegos_olimpicos = atletas_juegos_olimpicos.str.split(",", expand=True)
atletas_juegos_olimpicos
```

```python
atletas_juegos_olimpicos.columns = ['Apellido', 'Nombre', 'Vacio']
atletas_juegos_olimpicos[['Apellido', 'Nombre']]
```

**3. Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?**

```python
medallas = indice_año[["Country", "Medal"]]
medallas.head()
```

```python
medallas_por_pais = medallas.groupby(['Country']).count()
medallas_por_pais.sort_values("Medal", ascending=False)
```


**4. Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.**
