**Ejercicio 3**

```python
cant_registros = 48842

def porcentaje(n):
  return (n/cant_registros) * 100
```

Proporción de salarios respecto del total para personas de distinto sexo

```python
df['Sex'].groupby(df['Target']).value_counts(dropna=False).apply(porcentaje)
```

Proporción de salarios respecto del total para personas de distinta raza

```python
df['Race'].groupby(df['Target']).value_counts(dropna=False).apply(porcentaje)
```

Proporción de salarios respecto del total para personas de distinta educación

```python
df['Education'].groupby(df['Target']).value_counts(dropna=False).apply(porcentaje)
```
