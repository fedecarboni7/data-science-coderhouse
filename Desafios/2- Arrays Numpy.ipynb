**2. Recordar los ejercicios con funciones para cálculo de factorial y suma de serie de la Clase 02. Repetir ambos ejercicios, pero ahora utilizar las nuevas operaciones aprendidas con los ndarrays.**

Función factorial

```python
def factorial(num):
  if (num <= 1): return 1
  npa = np.array(range(1, num + 1)) # Creo un array desde 1 hasta el valor dado
  resultado = 1
  for i in npa:
    resultado *= i # Recorro el array, multiplicando cada número con el resultado de la multiplicación anterior
  return resultado
```

```python
factorial(5)
```

Función suma serie

```python
def suma_serie(min, max):
  if max <= min: return print("El valor de max tiene que ser mayor a min")
  npa = np.array(range(min, max + 1)) # Creo un array desde el valor mínimo hasta el valor máximo dados
  return npa.sum() # Devuelvo la suma de los valores dentro del array
```

```python
suma_serie(1, 3)
```
