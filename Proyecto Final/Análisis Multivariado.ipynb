```python
# Importamos librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lectura del dataset

df = pd.read_csv('/content/FIFA22_official_data.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df_filtrada = df[['Name', 'Age','Nationality','Overall',
       'Potential', 'Club','Value', 'Wage', 'Special','International Reputation','Skill Moves', 'Work Rate','Height', 'Weight','Best Position', 'Best Overall Rating']]
df_filtrada.head()
```

```python
df_filtrada['International Reputation'].unique()
```

```python
df_filtrada['Skill Moves'].unique()
```

```python
for index in range(0, len(df_filtrada['Value'])):
  
  st=str(df_filtrada['Value'].loc[index])
  replace_euro = st.replace('€','')
  df_filtrada['Value'].loc[index] = replace_euro

df_filtrada['Value']
```

```python
plt.rcParams['figure.figsize'] = (15, 10)

sns.heatmap(df_filtrada.corr(), annot = True, cmap = 'Wistia', fmt = ".2f")

plt.title('Heatmap for the Dataset', fontsize = 20)
plt.show()
```

Como se puede observar del Heatmap, las variables con una correlación mas fuerte son ('Skill Moves', 'Special'); ('Potential'/'Special', 'Best Overall Rating'); ('International Reputation','Best Overall Rating'). Estas relaciones son las que se analizarán mediante los gráficos siguientes.

```python
g = sns.FacetGrid(df_filtrada, hue='International Reputation', height = 5, aspect=1)
g.map(sns.scatterplot, 'Overall', 'Age')
plt.legend()
plt.show()
```

Acá se puede observar cuanto mayor es la reputación internacional, mayor es el Overall.

```python
g = sns.FacetGrid(df_filtrada, hue='International Reputation', height = 5, aspect=1)
g.map(sns.scatterplot, 'Overall', 'Potential')
plt.show()
```

En este gráfico se puede observar que a mejor reputación internacional, mayor el valor de potencial y por consiguiente mayor el Overall.

```python
g = sns.FacetGrid(df_filtrada, hue='Skill Moves', height = 5, aspect=1)
g.map(sns.scatterplot, 'Overall', 'Potential')
plt.show()
```

```python
g = sns.FacetGrid(df_filtrada, hue='Skill Moves', height = 5, aspect=1)
g.map(sns.scatterplot, 'Overall', 'Special')
plt.show()
```

En este gráfico podemos observar que a mayor Skill Moves, mayor el valor de Special.

```python
sns.pairplot(df_filtrada, hue='International Reputation')
```

En todos los gráficos, se remarca la tendencia lineal positiva en cuanto a reputación internacional frente a las demás variables.
