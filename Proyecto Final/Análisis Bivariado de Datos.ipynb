```python
datafiltrada = data[["Age","Overall","Potential","Skill Moves","Crossing","Finishing","Dribbling","Acceleration","Agility","Reactions","Vision","Stamina","SprintSpeed","International Reputation","DefensiveAwareness","Best Overall Rating"]]
datafiltrada.head()
```

```python
plt.rcParams['figure.figsize'] = (15, 10)

sns.heatmap(datafiltrada.corr(), annot = True, cmap = 'Wistia', fmt = ".2f")

plt.title('Heatmap for the Dataset', fontsize = 20)
plt.show()
```

---
**Relación entre EDAD y OVERALL**

```python
# tresbps vs target
plt.rcParams['figure.figsize'] = (12, 5)
sns.boxplot(x='Age', y='Overall', data=datafiltrada, palette = 'viridis')
plt.title('Relación entre Edad y Overall', fontsize = 20)
plt.show()
```

La relación entre Edad y Overall nos demuestra que generalmente los jugadores alcanzan un mayor nivel de sus capacidades a partir de los 23/24 años. Pero fuera de eso, no hay ningún otra diferencia causada principalmente por la edad.

---

**Relación entre AGILIDAD Y VELOCIDAD SPRINT**

```python
sns.scatterplot(x="Agility", y="SprintSpeed", data=data)
plt.title('Relación entre Agilidad y SprintSpeed', fontsize = 15, fontweight = 30)
```

Acá se nota claramente la relación lineal ascendente entre la agilidad y la velocidad sprint máxima de los jugadores, demostrando que a mayor agilidad, mayor velocidad
